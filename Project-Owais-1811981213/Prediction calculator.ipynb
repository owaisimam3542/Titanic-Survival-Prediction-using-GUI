```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import pandas as pd
import tkinter as tk

data=pd.read_csv("C:\\Users\hp\Desktop\\titanic.csv")
data=data.fillna(data.mean())
data=data.fillna(0)
X=data[['Pclass','Age','Fare']]
Y=data['Survived']
logreg = LogisticRegression()

logreg.fit(X, Y)
clf = RandomForestClassifier(n_estimators=10, random_state=0)

# Train the classifier
clf.fit(X,Y)
model = GaussianNB()
model.fit(X,Y)

classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X,Y)

root=tk.Tk()

canvas1 = tk.Canvas(root, width = 500, height = 500)
canvas1.pack()

label1 = tk.Label(root, text='PClass')
canvas1.create_window(100, 100, window=label1)
entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)


label2 = tk.Label(root, text='Age')
canvas1.create_window(100, 120, window=label2)
entry2 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 120, window=entry2)


label3 = tk.Label(root, text='Fare')
canvas1.create_window(100, 140, window=label3)
entry3 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 140, window=entry3)



def values():
    global a1 #our 1st input variable
    a1 = int(entry1.get()) 
    global a2 #our 1st input variable
    a2 = int(entry2.get())
    global a3 #our 1st input variable
    a3 = int(entry3.get())
  
    
    Prediction_result  = ('Prediction of Survival: ', logreg.predict([[a1,a2,a3]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='blue')
    canvas1.create_window(270, 400, window=label_Prediction)
    Prediction_result_1  = ('Prediction of Survival: ', clf.predict([[a1,a2,a3]]))
    label_Prediction_1 = tk.Label(root, text= Prediction_result_1, bg='blue')
    canvas1.create_window(270, 400, window=label_Prediction_1)
    Prediction_result_2  = ('Prediction of Survival: ', model.predict([[a1,a2,a3]]))
    label_Prediction_2 = tk.Label(root, text= Prediction_result_2, bg='blue')
    canvas1.create_window(270, 400, window=label_Prediction_2)
    Prediction_result_3  = ('Prediction of Survival: ', classifier.predict([[a1,a2,a3]]))
    label_Prediction_3 = tk.Label(root, text= Prediction_result_3, bg='blue')
    canvas1.create_window(270, 400, window=label_Prediction_3)
    
button1 = tk.Button (root, text='Logistic Regression',command=values, bg='blue') # button to call the 'values' command above 
canvas1.create_window(270, 200, window=button1)
    
button2 = tk.Button (root, text='Random Forest',command=values, bg='blue') # button to call the 'values' command above 
canvas1.create_window(270, 250, window=button2)

button3 = tk.Button (root, text='Naive Bias',command=values, bg='blue') # button to call the 'values' command above 
canvas1.create_window(270, 300, window=button3)

button4 = tk.Button (root, text='SVM',command=values, bg='blue') # button to call the 'values' command above 
canvas1.create_window(270, 350, window=button4)

root.mainloop()
```
